```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('events.csv')
```

```python
match_df = df[['match_id', 'match_name']].drop_duplicates()
teams = df[['match_id', 'is_home', 'team_id']].drop_duplicates()
goals = df.groupby(['match_id', 'is_home']).sum()['goals_scored']

for home_away, is_home in zip(['home', 'away'], [True, False]):
    mask = teams['is_home'] == is_home
    match_df[f'{home_away}_team_id'] = match_df.merge(teams.loc[mask, ['match_id', 'team_id']], on=['match_id'], how='left')[
        'team_id']

    mask = goals.index.get_level_values('is_home') == is_home
    match_df[f'{home_away}_goals'] = match_df.merge(goals[mask], on='match_id', how='left')['goals_scored']

match_df = match_df[['match_name', 'home_team_id', 'away_team_id', 'home_goals', 'away_goals']]
```
